```python
df = df.dropna(subset=['Engine Fuel Type'])
```

```python
df = df.dropna(subset=['Number of Doors'])
```

```python
df = df.fillna(df['Engine HP'].mean())
```

```python
df = df.fillna(df['Engine Cylinders'].mode())
```

```python
df['Market Category'].fillna(method = 'ffill', inplace = True)
```

```python
df.isnull().sum()
```

```python
# Converting to HTML
```

```python
import matplotlib.pyplot as plt
import base64
from io import BytesIO

fig = plt.figure()
#plot sth
plt.plot(df2['Popularity'])
tmpfile = BytesIO()
fig.savefig(tmpfile, format='png')
encoded = base64.b64encode(tmpfile.getvalue()).decode('utf-8')

html = 'Popularity' + '<img src=\'data:image/png;base64,{}\'>'.format(encoded) 

with open('popularity.html','w') as f:
    f.write(html)
```
